-----------------DATA TYPES------------

Data types in python - In programming, data type is an important concept.

Variables can store data of different types, and different types can do different things.

Python has the following data types built-in by default, in these categories:

```python
#Text type
a = "Hello"    #str (stores text data and to store characters we use ' ' or " " or """ """. )
print(type(a))
```

```python
#Numeric type
x = 5               #int ( stores whole numbers )
print(type(x))

y = 2.0             #float  ( stores decimal/real numbers )
print(type(y))

z = 1j              #complex  ( stores complex numbers )
print(type(z))  
```

```python
#Sequence type
b = ["Hello",56]        #LIST (ordered , Mutable,allows duplicates)
print(type(b))

c = ("apple","banana","cherry")     #TUPLE(ordered, immutable, allows duplicates)
print(type(c))

d = range(5)        #RANGE(generates sequence of numbers)
print(type(d))
```

```python
#Mapping type
e = {                           #dict ( it stores data in key value pairs.)
    "name" : "Arsh",
    "class" : "btech"
}
print(e["name"])
print(e.keys())
print(e.values())
print(e)
print(type(e))
```

```python
#Set Types
#set                                   (unordered, no duplicates, mutable)
f = {"Hello", 56 , "no"}
print(type(f))

#frozenset
g = frozenset({"Hello", 56 , "no"})         #its an immutable version of set
print(type(g))
```

```python
#Boolean
h = True            #stores two values ( True or False )
print(type(h))
```

```python
#Binary Types

#Byte
i = b"hello"            #immutable binary data
print(i)
print(type(i))

#Bytearray
j = bytearray(5)        #mutable binary data
print(j)
print(type(j))

#Memoryview
k = memoryview(bytes(5))        #allows access to binary objects memory
print(k)
print(type(k))
```

```python
#Nonetype
x = None
print(x)
print(type(x))
```

---PYTHON CASTING---

There may be times when you want to specify a type on to a variable. This can be done with casting. Python is an object-orientated language, and as such it uses classes to define data types, including its primitive types.

Casting in python is therefore done using constructor functions:

int() - constructs an integer number from an integer literal, a float literal (by removing all decimals), or a string literal (providing the string represents a whole number)

float() - constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer)

str() - constructs a string from a wide variety of data types, including strings, integer literals and float literals.

Examples given below:

```python
#int
x = int(1.0)
y = int(2)
z = int(3)

print(x,y,z)

#float
x = float(1.0)
y = float(2)
z = float(3)

print(x,y,z)

#str
x = str(1.0)
y = str(2)
z = str(3)
```
